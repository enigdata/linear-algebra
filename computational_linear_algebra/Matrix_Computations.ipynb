### Matrix and Tensor Products

#### Matrix-Vector Products

The matrix below gives the probabilities of moving from 1 health state to another in 1 year. If the current health states for a group are:

- 85% asymptomatic
- 10% symptomatic
- 5% AIDS
- 0% death

```python
transition_matrix = np.array(
    [[0.9,0.07,0.02,0.01],[0,0.93,0.05,0.02],[0,0,0.85,0.15],[0,0,0,1]]
)
```

```python
transition_matrix
```

```python
pi = np.array([0.85,0.1,0.05,0])
pi
```

```python
pi.shape
```

```python
pi = pi.reshape((4,1))
pi
```

```python
transition_matrix.T.dot(pi)
```

#### Matrix-Matrix Products



```python
Q = np.array(
    [[6,5,3,1],[3,6,2,2],[3,4,3,1]]
)

P = np.array(
    [[1.5,1],[2,2.5],[5,4.5],[16,17]]
)
```

```python
Q
```

```python
P
```

```python
Q@P
```
